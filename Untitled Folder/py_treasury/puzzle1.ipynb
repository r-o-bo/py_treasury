# Objective: Find the phone number

```python
# Good work on unzipping the file!
# You should now see 5 folders, each with a lot of random .txt files.
# Within one of these text files is a telephone number formated ###-###-#### 
# Use the Python os module and regular expressions to iterate through each file, open it, and search for a telephone number.
# Good luck!
```

```python
import os
import re
file_path = "c:\\Users\\ADMIN\\Downloads\\extracted_content"
phone_pattern = re.compile(r"(\d{3})-(\d{3})-(\d{4})")
```

```python
# Main Logic 

for folder, sub_folders, files in os.walk(file_path):
    print(f"Currently looking at {folder}")
    print("\nThe subfolders are:")
    for sub_fold in sub_folders:
        print(f"Subfolder: {sub_fold}")
    print("\nThe files are:")
    for f in files:
        print(f"File {f}")
        file_path = os.path.join(folder, f)
        try:
            with open(file_path, 'r') as file:
                contents = file.read()
                results = re.search(phone_pattern, contents)
                
                if results:
                    print(f"The phone number is in file {f}")
                    print(f"Phone number: {results.group()}")
                    break
        
                else:
                    print("No Match found, continuing search...")
        except Exception as e:
            print(f"Could not read file {f} due to {e}")
    print("\n")
            


        
        

        

        
   
```
