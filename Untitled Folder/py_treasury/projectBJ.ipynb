```python
import random
suits = ("Hearts", "Diamonds", "Spades", "Clubs")
ranks = ("Two", "Three", "Four", "Five", "Six", "Seven", "Eight", "Nine", "Ten", "Jack", "Queen", "King", "Ace")
values = {"Two":2, "Three":3, "Four":4, "Five":5, "Six":6, "Seven":7, "Eight":8, 
            "Nine":9, "Ten":10, "Jack":10, "Queen":10, "King":10, "Ace":11}

playing = True
            
```

```python
# CLASS CARD
class Card:
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        
    def __str__(self):
        return self.rank +  " of "   + self.suit
```

```python
# CLASS DECK
class Deck:
    
    def __init__(self):
        self.deck = []  # empty list
         # Iterating through the suit and rank tuples and attempting to create all 52 cards...
        for suit in suits:
            for rank in ranks:
                # Card object will be created here
                created_card = Card(suit,rank)
                self.deck.append(created_card)
    
    def __str__(self):
        # deck_c is used to accumulate the string representations of all the cards 
        deck_c = ''  
        for card in self.deck:
            # \n for readability
            deck_c += '\n ' + card.__str__() 
        return 'The deck has:' + deck_c


    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        dealt_card = self.deck.pop()
        return dealt_card
        
```

```python
# CLASS HAND 
class Hand:
    def __init__(self):
        self.cards = []  # start with an empty 
        self.value = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of aces
    
    def add_card(self,card):
        # This is taken from Deck.deal()
        self.cards.append(card)
        # card.rank is the key to dict values
        self.value += values[card.rank]
        if card.rank == "Ace":
            # Will be treated as boolean
            self.aces += 1
       
    
    def adjust_for_ace(self):

        # Check if over 21 and ace exists
        # Changes Ace value from (11 ---> 1)
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
```

```python
# CLASS CHIPS
class Chips:
    
    def __init__(self):
        self.total = 1000  # This can be set to a default value or supplied by a user input
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
    
    def lose_bet(self):
        self.total -= self.bet
```

```python
def take_bet(chips):
    while True:
        try:
            chips.bet = int(input("Please enter the amount you want to bet: "))
        except ValueError:
            print("Please enter an integer")
            
        else:
            if chips.bet > chips.total:
                print(f"You do not have the said amount, you have {chips.total}")
            else:
                break
```

```python
def hit(deck,hand):
    hand.add_card(deck.deal())
    hand.adjust_for_ace()
    
```

```python
def hit_or_stand(deck,hand):
    global playing  # to control an upcoming while loop

    while True:
        reply =  input("Do you want to hit or stand? (Enter h or s): ")
        if reply[0].lower() == "h":
            hit(deck,hand)
        elif reply[0].lower() == "s":
            print("Player stands..., Dealer will continue")
            playing = False
        else:
            print("Invalid response, please try again...")
            continue
        break
        
```
