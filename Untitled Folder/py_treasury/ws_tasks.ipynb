```python
# TASK: Import any libraries you think you'll need to scrape a website.
import requests
import bs4
```

```python
# TASK: Use requests library and BeautifulSoup to connect to http://quotes.toscrape.com/ and get the HTML text from the homepage.
res = requests.get("https://quotes.toscrape.com/")
soup = bs4.BeautifulSoup(res.text,"lxml")
soup
```

```python
# TASK: Get the names of all the authors on the first page.
authors = set()

for author in soup.select(".author"):
    authors.add(author.text)
print(authors)

    
```

```python
# TASK: Create a list of all the quotes on the first page.
quotes_list = []
for quotes in soup.select(".text"):
    quotes_list.append(quotes.text)
print(quotes_list)
```

```python
# TASK: Inspect the site and use Beautiful Soup to extract the top ten tags from the requests text shown on the top right from the home page 
for q in soup.select(".tag-item"):
    print(q.text)
```

```python
# TASK: Notice how there is more than one page, and subsequent pages look like this http://quotes.toscrape.com/page/2/.
# Use what you know about for loops and string concatenation to loop through all the pages and get all the unique authors on the website.
my_url = "https://quotes.toscrape.com/page/{}/"
all_authors = set()

try:
    for i in range(1,11):
        main_url = my_url.format(i) 
        res = requests.get(main_url)
        soup = bs4.BeautifulSoup(res.text,"lxml")
        peeps = soup.select(".author")

        for j in peeps:
            all_authors.add(j.text)
    print(all_authors)

            
    
except:
    print("Unusual error!")

finally:
    print("Program execution complete")

    
    

```
